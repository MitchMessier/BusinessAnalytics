## Prepare Data

➡️ Assignment Tasks
- Convert the Total_Amount into a number (not a string)
- Remove rows with empty (NULL) values
- Optional: Locate outliers in any of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
df['total_amount'] = df['total_amount'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.dropna(inplace=True)
```

```python
df = df.drop_duplicates()
```

## Summarize Data

➡️ Assignment Tasks
- Display the average price per month
- Display the total revenue for all months

```python
#average price
df['product_price'] = df['product_price'].replace('[\$,]', '', regex=True).astype(float)
average_price_per_month = df.groupby('order_month')['product_price'].mean()

average_price_per_month
```

```python
average_price_per_month = average_price_per_month.reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='order_month', y='product_price', data=average_price_per_month)
plt.title('Average Product Price Per Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.show()
```

```python
#total revenue
total_revenue_per_month = df.groupby('order_month')['total_amount'].sum()

total_revenue_per_month
```

```python
total_revenue_per_month = total_revenue_per_month.reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(total_revenue_per_month['order_month'], total_revenue_per_month['total_amount'], marker='o')
plt.title('Total Revenue Per Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```
