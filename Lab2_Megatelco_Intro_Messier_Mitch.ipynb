## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2)
```

## Income Distribution Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_clean['income'], kde=True)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

## Handset Price by Leave Status Box Plot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='leave', y='handset_price', data=df_clean)
plt.title('Handset Price by Leave Status')
plt.xlabel('Leave Status')
plt.ylabel('Handset Price')
plt.show()
```

## Average Over 15 Mins Calls per Month by Leave Status Bar Chart

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='leave', y='over_15mins_calls_per_month', data=df_clean)
plt.title('Average Over 15 Mins Calls per Month by Leave Status')
plt.ylabel('Average Over 15 Mins Calls per Month')
plt.xlabel('Leave Status')
plt.show()
```
